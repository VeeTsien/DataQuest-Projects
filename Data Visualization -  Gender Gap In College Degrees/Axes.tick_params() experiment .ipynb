```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.arange(0.0, 100.0, 1)
s = np.sin(2 * np.pi * t) * np.exp(-t * 0.01)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set_xticks(ticks = t, minor = True)
ax.set_yticks(ticks = s, minor = True)
ax.tick_params(which='both', width=2, color = 'b')

plt.show()
```

```python
t = np.arange(0.0, 100.0, 1)
s = np.sin(2 * np.pi * t) * np.exp(-t * 0.01)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.tick_params(which='major', length=7, color = 'g')

plt.show()
```

```python
t = np.arange(0.0, 100.0, 1)
s = np.sin(2 * np.pi * t) * np.exp(-t * 0.01)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set_xticks(ticks = t, minor = True)
ax.set_yticks(ticks = s, minor = True)
ax.tick_params(axis = 'x', which='minor', length=4, color='r')

plt.show()
```
