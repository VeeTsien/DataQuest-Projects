# Visualize Data and Explore 

### Goal:
* Quickly create plots using pandas methods 
* Explore features in the data and correlations between features with visualization

**Dictionary:**

`Rank` - Rank by median earnings (the dataset is ordered by this column).

`Major_code` - Major code.

`Major` - Major description.

`Major_category` - Category of major.

`Total` - Total number of people with major.

`Sample_size` - Sample size (unweighted) of full-time.

`Men` - Male graduates.

`Women` - Female graduates.

`ShareWomen` - Women as share of total.

`Employed` - Number employed.

`Median` - Median salary of full-time, year-round workers.

`Low_wage_jobs` - Number in low-wage service jobs.

`Full_time` - Number employed 35 hours or more.

`Part_time` - Number employed less than 35 hours.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Drop rows contain any missing values 
recent_grads.dropna(axis = 0, how = 'any', inplace = True)
```

## Create scatter plots to explore relations between features

### Fig. 1.1 Samples_size & Median relation

```python
# Samples_size & Median relation
recent_grads.plot.scatter(x = 'Sample_size', y = 'Median', title = 'Relation between Sample Size & Median')
```

### Fig. 1.2 Sample_size & Unemploymnet_rate relation

```python
# Sample_size & Unemploymnet_rate relation
recent_grads.plot.scatter(x = 'Sample_size', y = 'Unemployment_rate', title = 'Relation between Sample Size & Unemployment Rate')
```

### Fig. 1.3 Full_time & Median relation

```python
# Full_time & Median relation
recent_grads.plot.scatter(x = 'Full_time', y = 'Median', title = 'Relation between Full Time & Median')
```

### Fig. 1.4 ShareWomen & Unemployment_rate relation 

```python
# ShareWomen & Unemployment_rate relation 
recent_grads.plot.scatter(x = 'ShareWomen', y = 'Unemployment_rate', title = 'Relation between ShareWomen & Unemployment_rate')
```

### Fig. 1.5 Men & Median relation

```python
# Men & Median relation
recent_grads.plot.scatter(x = 'Men', y = 'Median', title = 'Relation between Men & Median')
```

### Fig. 1.6 Women & Median relation

```python
# Women & Median relation
recent_grads.plot.scatter(x = 'Women', y = 'Median', title = 'Relation between Women & Median')
```

### From the scatter plots above, we can conclude:
* Grads in majors with more students don't neccesarily make more money.

* Students that majored in female majority subjects don't neccesarily make more money.

* Just from Fig. 1.3 it's hard to tell if there's a correlation between full-time employees and median salary. 

* Among the 6 plots, the data is heavily right-skewed(the data in the 5 plots has a high variance near 0) for the x-axis except for Fig. 1.4 We will need to do some data manipulation to further explore the correlations.

## Create Scatter Matrix Plot

```python
from pandas.plotting import scatter_matrix
```

### Fig. 3.1 Scatter matrix plot from Sample_size and Median

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (10,10))
```

### Fig. 3.2 Scatter matrix plot from Sample_size, Median and Unemployment_rate

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (15,15))
```

To explore questions that were not answered in the last few steps, we will need to create more scatter matrix plots:

### Fig. 3.3 Scatter matrix plot from Median and ShareWomen

```python
scatter_matrix(recent_grads[['Median', 'ShareWomen']], figsize = (10,10))
```

### Fig. 3.4 Scatter matrix plot from Full_time and Median

```python
scatter_matrix(recent_grads[['Full_time', 'Median']], figsize = (10,10))
```

### From the plots above, to find anwers to these questions:
*Do students in more popular majors make more money?*

* From fig. 3.1, majors that makes the least and most money are not popular majors. Most majors fall into the common salary range. 

*Do students that majored in subjects that were majority female make more money?*

* From fig. 3.3, a trend shows as share of women goes up, the median goes down. Which means the opposite: students that majored in subjects that were majority female make less money.

*Is there any link between the number of full-time employees and median salary?*

* From fig. 3.4, grats that makes the least and most money have a smaller fulltime number. Most fall into the common salary range regardless of the fulltime number. 

## Explore Data With Bar Plots

### Fig. 4.1 Percentage of women in first ten rows

```python
recent_grads[:10].plot.bar(x = 'Major', y = 'ShareWomen')
```

### Fig. 4.2 Unemployment rate in first and last ten rows

```python
# Concatenate first 10 rows and last 10 rows along the index, then plot bar plot
pd.concat([recent_grads[:10],recent_grads[-10:]]).plot.bar(x = 'Major', y = 'Unemployment_rate')
```
